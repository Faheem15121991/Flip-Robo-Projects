# Assignment number 1: Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

# i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
teams=[]
matches=[]
points=[]
ratings=[]
```

```python
team=soup.find_all('span',class_='u-hide-phablet')
for i in team:
    teams.append(i.text)
    teams=teams[:10]
```

```python
match_top_row=soup.find_all('td',class_='rankings-block__banner--matches')
    #I observed that top row style is different and remaining rows are different
for i in match_top_row:
    matches.append(i.text)
# For remaining 9 teams
match=soup.find_all('td',class_='table-body__cell u-center-text')
# I am using even indices only for Matches. Will use odd ones for Points.
for i in range(0,len(match),2):
    matches.append(match[i].text)
    matches=matches[:10]
```

```python
point_top_row=soup.find_all('td',class_='rankings-block__banner--points')
#Similarly, I observed that top row style is different and remaining rows are different
for i in point_top_row:
    points.append(i.text)
# For remaining 9 teams
point=soup.find_all('td',class_='table-body__cell u-center-text')
# I am using odd indices only for Matches. Will use odd ones for Points.
for i in range(1,len(match),2):
    points.append(point[i].text)
    points=points[:10]
```

```python
rating=soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
#Similarly, I observed that top row style is different and remaining rows are different
for i in rating:
    ratings.append(i.text.replace('\n','').replace(' ',''))
# For remaining 9 teams
rating=soup.find_all('td',class_='table-body__cell u-text-right rating')
for i in rating:
    ratings.append(i.text.replace('\n','').replace(' ',''))
    ratings=ratings[:10]
```

```python
import pandas as pd
teams_odi=pd.DataFrame({})
teams_odi['teams']=teams
teams_odi['matches']=matches
teams_odi['points']=points
teams_odi['ratings']=ratings
```

```python
teams_odi
```

# Assignment no. 9. Write a python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.

```python
page=requests.get("https://www.monster.com/jobs/search/?q=Software-Developer&where=india")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
jobs=[]
Days=[]
Company=[]
Location=[]
```

```python
job=soup.find_all('h2',class_='title')
for i in job:
    jobs.append(i.text.replace('\r\n',''))
```

```python
day=soup.find_all('div',class_='meta flex-col')
for i in day:
    Days.append(i.text.replace('\n','').replace('AppliedSaved',''))
```

```python
comp=soup.find_all('span',class_='name')
# Even Indices are companies
for i in range(0,len(comp),2):
    Company.append(comp[i].text)
```

```python
# Odd Indices are locations
for i in range(1,len(comp),2):
    Location.append(comp[i].text.replace('\r\n',''))
```

```python
import pandas as pd
monster_jobs=pd.DataFrame({})
monster_jobs['jobs']=jobs
monster_jobs['Days']=Days
monster_jobs['Company']=Company
monster_jobs['Location']=Location
```

```python
monster_jobs
```
