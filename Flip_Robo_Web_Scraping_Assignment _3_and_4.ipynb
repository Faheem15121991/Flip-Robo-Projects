# Assignment no. 3- Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
# importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
Movie_Name=[]
IMDB_Rating=[]
Year_of_Release=[]
```

```python
titles2=soup.find_all('td',class_="titleColumn")
for i in titles2:
        full_name=i.find('a').get_text()
        Movie_Name.append(full_name)
```

```python
Movie_Name
```

```python
len(Movie_Name)
```

```python
ratings=soup.find_all('td',class_="ratingColumn imdbRating")
ratings
```

```python
for i in ratings:
    IMDB_Rating.append(i.get_text().replace("\n"," "))
IMDB_Rating
```

```python
len(IMDB_Rating)
```

```python
year_of_release=soup.find_all('span',class_="secondaryInfo")
year_of_release
```

```python
for i in year_of_release:
    Year_of_Release.append(i.get_text().replace("\n"," "))
Year_of_Release
```

```python
len(Year_of_Release)
```

```python
print(len(Movie_Name),len(IMDB_Rating),len(Year_of_Release))
```

```python
import pandas as pd
movies=pd.DataFrame([])
movies['Movie_Name']=Movie_Name[0:100]
movies['IMDB_Rating']=IMDB_Rating[0:100]
movies['Year_of_Release']=Year_of_Release[0:100]
```

```python
movies
```

```python
movies.to_csv("Top_100_Bollywood_Movies_list_IMDB.csv")
```

# Assignment no. 4- Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
page2=requests.get("https://bookpage.com/reviews")
```

```python
soup2=BeautifulSoup(page2.content,'html.parser')
```

```python
Book_Name=[]
Author_Name=[]
Genre=[]
Book_Review=[]
```

```python
book_name=soup2.find_all('h4',class_="italic")
for i in book_name:
        full_name=i.find('a').get_text()
        Book_Name.append(full_name)
```

```python
Book_Name
```

```python
author_name=soup2.find_all('p', class_="sans bold")
author_name
```

```python
for i in author_name:
   
    Author_Name.append(i.get_text().replace("\n",""))
Author_Name
```

```python
genre=soup2.find_all('p', class_="genre-links hidden-phone")
genre
```

```python
for i in genre:
   
    Genre.append(i.get_text().replace("\n",""))
Genre
```

```python
book_review=soup2.find_all('p', class_="excerpt")
book_review
```

```python
for i in book_review:
   
    Book_Review.append(i.get_text().replace("\n",""))
Book_Review
```

```python
len(Book_Review)
```

```python
import pandas as pd
book=pd.DataFrame([])
book['Book_Name']=Book_Name
book['Author_Name']=Author_Name
book['Genre']=Genre
book['Book_Review']=Book_Review
```
