# Assignment number 1: Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
# importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page1=requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
soup1=BeautifulSoup(page1.content,'html.parser')
```

```python
Titles=[]
```

```python
titles=soup1.find_all(['h1', 'h2','h3'])
titles
```

```python
for i in titles:
   
    Titles.append(i.get_text().replace("\n",""))
Titles
```

```python
print(len(Titles))
```

```python
import pandas as pd
Wiki_Header_Tags=pd.DataFrame([])
Wiki_Header_Tags['Titles']=Titles
```

```python
Wiki_Header_Tags
```

```python
Wiki_Header_Tags.to_csv("wiki_all_header_tags.csv")
```

# Assignment no. 2- Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
page2=requests.get("https://www.imdb.com/chart/top/")
```

```python
soup2=BeautifulSoup(page2.content,'html.parser')
```

```python
Movie_Name=[]
IMDB_Rating=[]
Year_of_Release=[]
```

```python
titles2=soup2.find_all('td',class_="titleColumn")
for i in titles2:
        full_name=i.find('a').get_text()
        Movie_Name.append(full_name)
```

```python
Movie_Name[0:5]
```

```python
len(Movie_Name)
```

```python
ratings=soup2.find_all('td',class_="ratingColumn imdbRating")
ratings
```

```python
for i in ratings:
    IMDB_Rating.append(i.get_text().replace("\n"," "))
IMDB_Rating
```

```python
len(IMDB_Rating)
```

```python
year_of_release=soup2.find_all('span',class_="secondaryInfo")
year_of_release
```

```python
for i in year_of_release:
    Year_of_Release.append(i.get_text().replace("\n"," "))
Year_of_Release
```

```python
len(Year_of_Release)
```

```python
print(len(Movie_Name),len(IMDB_Rating),len(Year_of_Release))
```

```python
import pandas as pd
movies=pd.DataFrame([])
movies['Movie_Name']=Movie_Name[0:100]
movies['IMDB_Rating']=IMDB_Rating[0:100]
movies['Year_of_Release']=Year_of_Release[0:100]
```

```python
movies
```

```python
movies.to_csv("Top_100_Hollywood_Movies_list_IMDB.csv")
```
